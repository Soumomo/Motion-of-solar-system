```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import scienceplots
plt.style.use(['science','grid','no-latex'])

gal_lon,gal_lat,Temperature,cos_theta=np.loadtxt('final-data.csv',delimiter=',',unpack=True)
```

```python
dipole=Temperature-2.727780
dipole=dipole
dipole=dipole[1100:2300]
```

```python
dipole.size
```

```python
plt.figure(figsize=(10,4))
plt.plot(np.arange(dipole.size),dipole,'ko',ms=1)
```

```python
def cos_function(x,n,a,b):
    return a*np.cos(n*x-b)

popt,pcov=curve_fit(cos_function,dipole,np.arange(dipole.size),p0=[0.002,0.003,200])
```

```python
popt
```

```python
plt.figure(figsize=(10,4))
#plt.plot(np.arange(dipole.size),dipole,'ko',ms=1)
plt.plot(np.arange(dipole.size),cos_function(np.arange(dipole.size),0.002,0.003,550),'r-',ms=1)
plt.show()
```

```python
from scipy.constants import h,c,k
```

```python
def Temp_theta(T_cmbr,v,cos_theta):
    T_cmbr(1+(v*cos_theta)/c)
```

```python
chisq=[]
Temp=np.linspace(2.7,2.8,num=6067)
for  i in np.arange(Temperature.size):
    r = Temperature[i] - Temp_theta(Temp[i],320*1e3,cos_theta[i])
    chisq.append(sum((r)**2))
```
